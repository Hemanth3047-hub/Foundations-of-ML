```python
from sklearn import datasets
dataset = datasets.load_digits()#importing the data set using load_digits()
X = dataset.data#actual data
y = dataset.target#target
```

```python
from sklearn.manifold import TSNE
model = TSNE(n_components=2,perplexity=30,n_iter=250,random_state=0)#TSNE inbuilt from sklearn taking perplexity=30 and no_of_iterations=250
model_2d = model.fit_transform(X)#fit_transform
```

```python
target_ids = range(len(dataset.target_names))#storing the final target values
from matplotlib import pyplot as plt
plt.figure(figsize=(16,10))#converting the fig sixe into 16X10
colors = 'b', 'g', 'r', 'c', 'm', 'y', 'k', 'orange', 'navy', 'plum'#assigning the colors for each digit 
for i, c, label in zip(target_ids, colors, dataset.target_names):
    plt.scatter(model_2d[y == i, 0], model_2d[y == i, 1], c=c, label=label)#scattering (plotting the digits)
plt.legend()
plt.show()
```
